```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
    return (x+3)**2
```

```python
def grad(x):
    return 2 * (x+3)
```

```python
def gradient(initial_x, learning_rate, iterations):
    x = initial_x
    x_history = [x]

    for i in range(iterations):
        gradient = grad(x)
        x = x - learning_rate * gradient
        x_history.append(x)
    return x, x_history

initial_x=2
learning_rate=0.1
iterations=100

min_x, x_history = gradient(initial_x, learning_rate, iterations) 
```

```python
x_values = np.linspace(-10, 5, 400)
y_values = func(x_values)
```

```python
plt.figure(figsize=(12,5))
plt.plot(x_values, y_values, label='y=(x+3)**2')
plt.scatter(x_history, func(np.array(x_history)), color='red', label='Gradient Descent Steps')
plt.plot(x_history, func(np.array(x_history)), color='red', linestyle='--')
plt.xlabel='x'
plt.ylabel='y'
plt.title('Gradient Descent Optimization')
plt.legend()
plt.grid(True)
plt.show()
print(f"Local minima occurs at x = {min_x}")
```
