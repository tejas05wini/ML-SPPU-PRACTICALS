```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Load Dataset
cancer_dataset = load_breast_cancer()
df = pd.DataFrame(cancer_dataset.data, columns=cancer_dataset.feature_names)

print("Dataset Shape:", df.shape)
df.head()
```

```python
# Standardization
scaler = StandardScaler()
scaled_data = scaler.fit_transform(cancer_dataset.data)
```

```python
# Apply PCA (Reduce to 2 Components)
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

print("\nExplained Variance (2 Components):", pca.explained_variance_)
```

```python
# Scatter Plot of First Two Principal Components
plt.figure(figsize=(8,6))
plt.scatter(pca_data[:,0], pca_data[:,1], c=cancer_dataset.target, cmap='plasma')
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
plt.title("PCA on Breast Cancer Dataset")
plt.show()
```

```python
# PCA with All Components (to see explained variance distribution)
pca_all = PCA()
pca_all.fit(scaled_data)
print("\nExplained Variance (All Components):")
print(pca_all.explained_variance_)
```
